```python
# Preprocessing

def preprocessing(data, flag=False):
  # Lowercasing
  data= data.lower()
  if flag:
    print()
    print("Lower cased data: ", data)

  # Tokenization
  # tokenizer = nltk.WordPunctTokenizer()
  # data= tokenizer.tokenize(data)
  data= word_tokenize(data)
  if flag:
    print()
    print("Tokenized data: ", data)

  # Non stop words
  stopWords= stopwords.words("english")

  tempList= []
  for j in data:
    if(j not in stopWords):
      tempList.append(j)
  data= copy.deepcopy(tempList)

  if flag:
    print()
    print("Non stop words: ", data)

  # Punctuation removal
  puncts= string.punctuation
  tempList= []

  for j in data:
    if(j.isalpha()):
      tempList.append(j)
    else:
      tempList.append("")
  data= copy.deepcopy(tempList)

  if flag:
    print()
    print("Punctuation removed: ", data)

  # White space removal
  tempList= []
  for j in data:
    if(j!=""):
      tempList.append(j)
  data= copy.deepcopy(tempList)

  if flag:
    print()
    print("White space removed: ", data)

  if flag:
    print()
    print()
    print()

  return data
```

# Unigram Inverted Index and Boolean Queries

```python
# Creating Unigram Inverted Index
uii= {}

for i in range(len(preprocessedData)):
  for j in range(len(preprocessedData[i])):
    if preprocessedData[i][j] not in uii.keys():
      temp= [i+1]
      uii[preprocessedData[i][j]]= temp
    elif uii[preprocessedData[i][j]][-1]!=i+1:
      uii[preprocessedData[i][j]].append(i+1)

# Printing Index
for key, value in uii.items():
  print(key, value)
```

```python
import pickle

with open('UnigramInvertedIndex.pickle', 'wb') as i:
    pickle.dump(uii, i)
with open('UnigramInvertedIndex.pickle', 'rb') as i:
    uii = pickle.load(i)
```

```python
def AND(a, b):
  res= []

  m= len(a)
  n= len(b)

  a.sort()
  b.sort()

  i= 0
  j= 0

  while(True):
    if(i>=m or j>=n):
      break
    if a[i]<b[j]:
      i+=1
    elif a[i]>b[j]:
      j+=1
    else:
      res.append(a[i])
      i+=1
      j+=1

  return res
```

```python
def OR(a, b):
  res= []

  m= len(a)
  n= len(b)

  res= a
  for i in b:
    if i not in res:
      res.append(i)

  return res
```

```python
def AND_NOT(a, b):
  # print(len(b))
  b1= [i for i in range(1, 1000) if i not in b]
  # print(len(b1))
  return AND(a,b1)
```

```python
def OR_NOT(a, b):
  b1= [i for i in range(1, 1000) if i not in b]
  return OR(a,b1)
```

```python
# Genralised Query
def genQuery():
  t= int(input("Enter the number of queries: "))
  for l in range(t):
    print()
    with open('UnigramInvertedIndex.pickle', 'rb') as i:
      uii = pickle.load(i)
    inpSeq= input("Enter the sequence: ")
    operations= input("List of operations: ")

    # inpSeq pe preprocessing
    inpList= preprocessing(inpSeq)

    # operations to list
    opsList= operations.split(", ")

    # using above functions process to output
    if inpList[0] not in uii.keys():
      res= []
    else:
      res= uii[inpList[0]]

    for i in range(len(opsList)):
      if inpList[i+1] not in uii.keys():
        temp= []
      else:
        temp= uii[inpList[i+1]]
      if opsList[i]=="AND":
        res= AND(res, temp)
      elif opsList[i]=="OR":
        res= OR(res, temp)
      elif opsList[i]=="AND NOT":
        # print("hello")
        res= AND_NOT(res, temp)
      elif opsList[i]=="OR NOT":
        res= OR_NOT(res, temp)
      # print(len(res))

    print("Query ", l, ": ", inpList[0], end=" ")
    for i in range(1, len(inpList)):
      print(opsList[i-1], inpList[i], end=" ")

    print()
    print("Number of documents retrieved for query ", l, ": ", len(res))
    print("Name of the documents retrieved for query ", l, ": ", end=" ")

    for i in range(len(res)):
      print("file", end="")
      print(res[i], end="")
      print(".txt", end= ", ")
```

# Positional Index and Phrase Queries

```python
# Creating Positional Index
pi= {}

for i in range(len(preprocessedData)):
  for j in range(len(preprocessedData[i])):
    if preprocessedData[i][j] not in pi.keys():
      temp= {i+1: [j+1]}
      pi[preprocessedData[i][j]]= temp
    else:
      exst= pi[preprocessedData[i][j]]
      if(i+1 in exst.keys()):
        pi[preprocessedData[i][j]][i+1].append(j+1)
      else:
        pi[preprocessedData[i][j]][i+1]= [j+1]

# Printing Index
for key, value in pi.items():
  print(key, value)
```

```python
import pickle

with open('PositionalIndex.pickle', 'wb') as p:
    pickle.dump(pi, p)
with open('PositionalIndex.pickle', 'rb') as p:
    pi = pickle.load(p)
```

```python
def phraseHelper(lst, pi):
  for i in lst:
    if i not in pi.keys():
      print(i, "not in any document")
      return []

  # Logic for Phrase Queries
  res= []
  ref= pi[lst[0]]
  # print("ref: ", ref)
  i= 1

  while(i<len(lst)):
    ref1= {}
    ref2= pi[lst[i]]
    # print("ref2: ", ref2)
    # ref aur ref2 ko compare karke ref update karna hai
    for i in ref.keys():
      if i in ref2.keys():
        # print(ref2[i])
        for j in range(len(ref2[i])):
          if ref2[i][j]==ref[i][0]+1:
            ref1[i]= ref2[i][j:]
    ref= ref1
    i+= 1

  for i in ref:
    res.append(i)

  return res
```

```python
def phraseQuery():
  t= int(input("Number of queries to execute: "))
  for l in range(t):
    print()
    s= input("Enter the phrase: ")
    lst= preprocessing(s)

    res= phraseHelper(lst, pi)
    print("Number of documents retrieved for query ", l,"using positional index: ", len(res))
    print("Names of documents retrieved for query ", l,"using positional index: ", end= "")
    for i in range(len(res)):
      print("file", end="")
      print(res[i], end="")
      print(".txt", end= ", ")
```
